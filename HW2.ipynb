E2 (c) Generate N = 10000 samples of Sn for n = 10, 100 and 10000 using a computer. Plot a histogram, and check graphically and numerically the central limit theorem.

```python
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import math
from random import choice

N=10000   # Number of samples
n=10      # Number of steps

# Define a class to realize the random walk
class RandomWalk:
    def __init__(self,n):
        self.n=n
        self.value=[]
    def walk(self):
        while len(self.value)<self.n:
            x=choice([1,-1])
            self.value.append(x)
        x_sum=sum(self.value)
        return x_sum

# Generate the random walk samples
samples=[]
i=1
while i<=N:
    rw=RandomWalk(n)     # Initialize a RandomWalk object rw
    samples.append(rw.walk())
    i=i+1
print("N=",len(samples)) # Check the number of samples we have generated
print("n=",n)            # Print the number of steps

# Draw the histogram
num_bins=20                         # The number of bins
nn,bins,patches=plt.hist(samples,num_bins)
plt.plot(bins)
plt.xlabel('Random walk data')    
plt.ylabel('Number of data in each bin')    
plt.title('10000 samples of arithmetic random walk')
plt.show()

# Calculate the (sample-n*mean)/(std*sqrt(n))
mean=0
std=1
n_mean=n*mean
std_d=std*math.sqrt(n)
store_value=[]
for item in samples:
    store_value.append((item-n_mean)/std_d)
    
# Check numerically the central limit theorem
print("Check numerically the central limit theorem:")
mean_sv=sum(store_value)/N

aa=[]
for item in store_value:
    x=pow((item-mean_sv),2)
    aa.append(x)
variance_sv=sum(aa)/(N-1)

print("mean:",mean_sv) # Print the mean
print("variance:",variance_sv) # Print the variance

# Check graphically the central limit theorem
print("\nCheck graphically the central limit theorem:")
num_bins2=20                        
nn2,bins2,patches2=plt.hist(store_value,num_bins2,normed=1, facecolor='blue',\     
alpha=0.5)
y = mlab.normpdf(bins2, 0, 1) # Fit a standard normal distribution line
plt.plot(bins2,y,'r--')
plt.xlabel('Random walk data')    
plt.ylabel('Probability')    
plt.title('10000 samples of arithmetic random walk (standardized)')
plt.show()

```
